```python
df_labels, label_encoder  = prepare_folder_structure(minority_balanced=False, test_size=0.2)
```

```python
df_labels.groupby(['is_train','label']).count()
```

```python
df_labels = minority_balance_dataframe_by_multiple_categorical_variables(df_labels, ['label'])
described, described_test = extract_features(df_labels=df_labels, save=False)
y_train_encoded = np_utils.to_categorical(df_labels[df_labels.is_train].label.apply(lambda x: label_encoder.transform([x])[0]).values)
y_train = df_labels[df_labels.is_train].label
described_test.shape
```

```python
rfc = RandomForestClassifier(n_estimators=200, oob_score=True, n_jobs=-1)
rfc.fit(described, y_train)
rfc.predict(described_test)
```

```python
from pandas_ml import ConfusionMatrix
cm = ConfusionMatrix(df_labels[~df_labels.is_train].label.values, rfc.predict(described_test))
cm.classification_report
```
